# Procesamiento de Lenguage Natural

## Taller #5: TF-IDF


```python
import re
import numpy as np
import pandas as pd

from nltk.corpus import stopwords
stopwords_sp = stopwords.words('spanish')

from nltk.stem.snowball import SnowballStemmer
spanishStemmer=SnowballStemmer("spanish")

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_distances
```

# Punto 1: Pre-Procesamiento

-  Leer el archivo `Princesas.csv` usando `pandas` y crear una nueva columna con el texto en minúscula, sin caracteres especiales ni números, sin palabras vacias y hacer stemming de las palabras

```python
# data
data = pd.read_csv('Princesas.csv')
data 
```

```python
# función
def pre_procesado(texto):
    texto = texto.lower()
    texto = re.sub(r"[\W\d_]+", " ", texto)
    texto = " ".join([palabra for palabra in texto.split() if palabra not in stopwords_sp])
    texto = " ".join([spanishStemmer.stem(palabra) for palabra in texto.split()])
    return texto
```

```python
# data procesada
data['pre-procesado'] = data['Personalidad'].apply(lambda texto: pre_procesado(texto))
data
```

# Punto 2: TF-IDF

-  Crear la matriz TF-IDF

```python
# tfidf_matrix
tfidf_vect = TfidfVectorizer() 
tfidfPrincesas = tfidf_vect.fit_transform(data['pre-procesado'].values) 

tfidf_matrix = pd.DataFrame(tfidfPrincesas.toarray(), columns=tfidf_vect.get_feature_names()) 
tfidf_matrix.index = data['Princesa'].values 

tfidf_matrix.T.round(3)
```

# Punto 3: Distancia del coseno
-  Calcular la distancia del coseno entre cada una de las princesas
-  ¿Cuáles son las princesas más parecidas?
-  ¿Cuáles son las princesas más diferentes?

```python
# dist_cos

dist_cos = cosine_distances(tfidf_matrix.values)
dist_cos = pd.DataFrame(dist_cos, columns = tfidf_matrix.T.columns, index = tfidf_matrix.T.columns)
dist_cos
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import colors
fig, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(dist_cos, mask=dist_cos <= 0.5,
            cmap='Blues', norm=colors.Normalize(vmin=0.8, vmax=1), vmin=0.8, vmax=1, cbar=False,
            annot=True, fmt='.2f', annot_kws={"size": 10},
            xticklabels=tfidf_matrix.T.columns, yticklabels=tfidf_matrix.T.columns, ax=ax)
plt.show()
```

¿Cuáles son las princesas más parecidas?
Bella y Aurora con una distancia de 0.83, seguido por Aurora y Blancanives (0.84) y cenicienta y  Blancanieves (0.85)

¿Cuáles son las princesas más diferentes?
Tiana y Mulan (0.99), Mérida y Blancanieves (0.99), Mérida y Mulan (0.99)
