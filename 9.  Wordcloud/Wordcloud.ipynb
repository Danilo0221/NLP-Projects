```python
import pandas as pd
 
data = pd.read_csv('desktop/archivos/hween_clean.csv')

data.head()
```

```python
import numpy as np

from wordcloud import WordCloud
from wordcloud import ImageColorGenerator

from nltk.corpus import stopwords
stopwords = stopwords.words('spanish')

import matplotlib.pyplot as plt
import PIL.Image
from IPython.display import display
```

```python
fantasma = PIL.Image.open("desktop/archivos/fantasma.jpg")

display(fantasma)
```

```python
mask = np.array(fantasma)

mask
```

```python
palabras = data.texto.str.cat(sep=' ')
font_path = 'Desktop/archivos/Halloween Day.ttf' #Fuente descargada de Dafont
wordcloud = WordCloud(random_state=1234, width=500, height=500,
                      max_font_size=200, max_words=150,
                      background_color='white', colormap='twilight',
                      stopwords=stopwords, font_path=font_path,
                      mask=mask, contour_width=0.5, contour_color='black').generate(palabras) # https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html
 
wordcloud.to_file("fantasma.jpg")


plt.figure(figsize=(12,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
